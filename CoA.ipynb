## Year, Municipality, Province

```python
import re
muni_re = re.compile("[\n\r].*MUNICIPALITY OF\s*([^\n\r]*)", flags=re.IGNORECASE)
municipality = muni_re.findall(text)

prov_re = re.compile("[\n\r].*PROVINCE OF\s*([^\n\r]*)", flags=re.IGNORECASE)
province = prov_re.findall(text)

year_re = re.compile("[\n\r].*FOR THE YEAR ENDED\s*([^\n\r]*)", flags=re.IGNORECASE)
year_ended = year_re.findall(text)
```

## Qualified/Unqualified Opinion

```python
qualified_re = re.compile("[\n\r].*qualified opinion\s*([^\n\r]*)", flags=re.IGNORECASE)
qualified = qualified_re.findall(text)

unqualified_re = re.compile("[\n\r].*unqualified opinion\s*([^\n\r]*)", flags=re.IGNORECASE)
unqualified = unqualified_re.findall(text)
```

```python
print('Qualified:', len(qualified),'Unqualified:', len(unqualified))
```

## Status of implementation

The code below assumes that the last table of the audit report is the status of implementation.

```python
import pandas as pd

table = document.tables[len(document.tables)-1]
df = pd.DataFrame()

# Data will be a list of rows represented as dictionaries containing each row's data.
data = []

keys = None
for i, row in enumerate(table.rows):
    text = (cell.text for cell in row.cells)

    # Establish the mapping based on the first row headers; these will become the keys of our dictionary
    if i == 0:
        keys = tuple(text)
        continue

    # Construct a dictionary for this row, mapping keys to values for this row
    row_data = dict(zip(keys, text))
    temp = pd.DataFrame(row_data, index=[i-1])
    df = df.append(temp)
```
